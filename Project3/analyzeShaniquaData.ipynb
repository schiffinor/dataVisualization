```python
import os
import random
import numpy as np
import matplotlib.pyplot as plt

import charts
import data
import linear_regression

plt.style.use(['seaborn-v0_8-colorblind', 'seaborn-v0_8-darkgrid'])
plt.rcParams.update({'font.size': 20})

np.set_printoptions(suppress=True, precision=5)

# Automatically reload external modules
%load_ext autoreload
%autoreload 2

tStep = 10000
xdata = np.linspace(0, 100, tStep)

LFNAmplitude = 0.5
LFFreq = 0.1
HFNAmplitude = 0.2
HFFreq = 1

def f(x):
    return (0.5*(x**3)+2*x**2-1.4*x+0) + ((LFNAmplitude *np.random.normal())* np.sin(2*np.pi*LFFreq*x))+((HFNAmplitude *np.random.normal())* np.sin(2*np.pi*HFFreq*x))

ydata = np.array([f(xi) for xi in xdata])

plt.plot(xdata, ydata)
plt.show()
```

```python
from data import Data
from linear_regression import RegressTypes

dataSet = np.hstack((xdata.reshape(-1, 1), ydata.reshape(-1, 1)))
print(dataSet.shape)
print(dataSet)
datas = Data(headers=['x', 'y'], data=dataSet, header2col={"x": 0, "y": 1}, )
lin_reg = linear_regression.LinearRegression(datas)

for i in range(1, 6):
    print(f"Linear Regression Degree {i}")
    lin_reg.pair_plot(datas.get_headers(), reg_type=RegressTypes.linear, degree= i, hists_on_diag=True)
    plt.show()
for i in range(1, 6):
    print(f"Polynomial Regression Degree {i}")
    lin_reg.pair_plot(datas.get_headers(), reg_type=RegressTypes.polynomial, degree= i, hists_on_diag=True)
    plt.show()
for i in range(1, 6):
    print(f"Exponential Regression Degree {i}")
    lin_reg.pair_plot(datas.get_headers(), reg_type=RegressTypes.exponential, degree= i, hists_on_diag=True)
    plt.show()
for i in range(1, 6):
    print(f"Sinusoidal Regression Degree {i}")
    lin_reg.pair_plot(datas.get_headers(), reg_type=RegressTypes.sinusoidal, degree= i, hists_on_diag=True)
    plt.show()

```
